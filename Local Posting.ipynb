```python
import bz2
import pickle
from collections import Counter
from pathlib import Path

# Path to the already downloaded pageview file
pv_path = '/Users/ofekzini/Documents/Data Engineering/Fall 2024/אחזור/Project/IR-Project/pageviews-202108-user.bz2'
p = Path(pv_path)
pv_name = p.name
pv_temp = f'{p.stem}-4dedup.txt'
pv_clean = f'{p.stem}.pkl'

# Filter for English pages and keep the relevant fields
# Create a Counter (dictionary) that sums up the page views for the first 2000 articles
wid2pv = Counter()
page_count = 0  # Counter to track the number of pages processed
max_pages = 2000  # Limit to 2000 pages

with bz2.open(pv_path, 'rt') as f_in, open(pv_temp, 'wt') as f_out:
    for line in f_in:
        if line.startswith("en.wikipedia"):
            parts = line.split()
            if len(parts) >= 6 and parts[2].isdigit() and parts[4].isdigit():
                # print(page_count)
                article_id = int(parts[2])
                page_views = int(parts[4])
                wid2pv.update({article_id: page_views})
                f_out.write(f"{article_id} {page_views}\n")
                page_count += 1

                if page_count >= max_pages:  # Stop after 2000 pages
                    break

# Write out the counter as a binary file (pickle it)
with open(pv_clean, 'wb') as f:
    pickle.dump(wid2pv, f)

# Read in the counter
with open(pv_clean, 'rb') as f:
    wid2pv = pickle.load(f)

# Example: Print the first 10 entries
print(list(wid2pv.items())[:10])
```
